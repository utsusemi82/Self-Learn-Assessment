```python
# Create a new EntitySet
es = ft.EntitySet(id="Ecommerce")

# Add the customer data
es = es.add_dataframe(dataframe_name="customers",
                      dataframe=customers,
                      index="CustomerID"
)

# Add the products data
es = es.add_dataframe(dataframe_name="products",
                      dataframe=products,
                      index="ProductID"
)

# Add the orders data
es = es.add_dataframe(dataframe_name="orders",
                      dataframe=orders,
                      index="OrderID"
)

# Add the order details data
es = es.add_dataframe(dataframe_name="order_details",
                      dataframe=order_details,
                      make_index=True,
                      index="OrderDetailsID"
)

# Add the customer feedback data
es = es.add_dataframe(dataframe_name="customer_feedback",
                      dataframe=customer_feedback,
                      index="FeedbackID"
)

es
```

### **Step 3: Establish Relationships between the Entities**

```python
es = es.add_relationship("customers", "CustomerID", "orders", "CustomerID")
es = es.add_relationship("orders", "OrderID", "order_details", "OrderID")
es = es.add_relationship("products", "ProductID", "order_details", "ProductID")
es = es.add_relationship("orders", "OrderID", "customer_feedback", "OrderID")
es = es.add_relationship("customers", "CustomerID", "customer_feedback", "CustomerID")
es = es.add_relationship("products", "ProductID", "customer_feedback", "ProductID")

es
```

### **Step 4: Perform Deep Feature Synthesis**

```python
# Develop DFS

#for orders DF
feature_matrix_orders, feature_defs_orders = ft.dfs(entityset=es,
                                      target_dataframe_name='orders',
                                      agg_primitives=['sum', 'mean', 'count'],
                                      trans_primitives=['month', 'day', 'year'],
                                      max_depth=2)

feature_matrix_orders.head(5)
```
